## 2 - Théorie: Mot-clé `assert` pour déboguer et plus d'exercices

`assert` n'est pas une fonction, mais un mot-clé du langage python. Il a pour sens : *si la condition donnée est fausse, stoppe le code et déclenche une erreur*.

```python
x = 10

y = 20

assert x + y < 25, "La somme n'est pas aussi petite que prévu"
```

En quoi est-ce que volontairement déclencher des erreurs peut-il nous être utile?

Parfois, mieux vaut que le code plante *tôt* quand la raison du plantage est claire, plutôt que *tard* quand on a déjà calculé un paquet de choses avec des variables qui ne faisaient pas de sens.

Supposons qu'on a la séquence de fonctions suivante :

```python
def addition_tableaux(u, v):
  """Additionne deux tableaux de dimensions quelconques"""

  somme = []

  for i in range(len(u)):
    somme.append(u[i] + v[i])

  return somme


def total(v):
  resultat = 0
  for x in v:
    resultat += x
  return resultat

def carre(x):
  return x*x


# Ça devrait faire [1+3, 2+4, 0+5] = [4, 6, 5]
somme_vecteurs = addition_tableaux([1,2], [3,4,5])

total_carre = carre(total(somme_vecteurs))

# Ça devrait faire : 15² = 225
display(total_carre)

# ... Pourquoi ça donne la mauvaise réponse?
```

Où est l'erreur?

Dans `addition_tableaux`, dans `total` ou dans `carre`?

Une des fonction faisait une hypothèse incorrecte avec ses données : `addition_vec` présume que les deux vecteurs auront la même taille, ce qui n'est pas le cas dans l'exemple d'utilisation de la méthode.

**`assert` permet de rendre explicite nos hypothèses de départ, en s'assurant que le code ne continue pas avec des données invalides si on se trompe.**

Même chose, avec un petit assert en début de fonction :

```python
def addition_tableaux(u, v):
  """Additionne deux tableaux de dimensions quelconques"""

  assert len(u) == len(v), "Les tableaux doivent avoir la même taille"

  somme = []

  for i in range(len(u)):
    somme.append(u[i] + v[i])

  return somme


def total(v):
  resultat = 0
  for x in v:
    resultat += x
  return resultat

def carre(x):
  return x*x


# Ça devrait faire [1+3, 2+4, 0+5] = [4, 6, 5]
somme_vecteurs = addition_tableaux([1,2], [3,4,5])

total_carre = carre(total(somme_vecteurs))

# Ça devrait faire : 15² = 225
print(total_carre)
```

✅ Le code plante *dès le assert*, sans continuer plus loin

### 2.1 le `produit_scalaire(u, v)`, qui retourne le produit scalaire (*dot product*) entre deux vecteurs à N dimensions

En 2D, le produit scalaire est :

$$(x, y) \cdot (a, b) = xa + yb$$

En 3D, ça serait plutôt :

$$||(x, y, z)|| = (x, y, z) \cdot (a, b, c) = xa + yb + zc$$

Généralisez la logique à n'importe quel nombre de dimensions.


### 2.2 le `module()` d'un vecteur à N dimensions

Le module en 2D est :

$$||(x, y)|| = \sqrt{x^2 + y^2}$$

En 3D ça serait :

$$\sqrt{x^2 + y^2 + z^2}$$

En n'importe quel nombre de dimensions, ça serait :

$$\sqrt{x^2 + y^2 + ... + w^2}$$


### 2.3 `vecteur_unitaire(v)`, qui retourne le vecteur où chaque élément est divisé par le module du vecteur

$$\vec{v}_\text{unitaire} = \frac{\vec{v}}{|| \vec{v} ||} = (\frac{v_0}{|| \vec{v} ||}, \frac{v_1}{|| \vec{v} ||}, ..., \frac{v_{N-1}}{|| \vec{v} ||}, )$$

### 2.4 `angle_entre_vecteurs(u, v)`

$$\vec{x} \cdot \vec{y} = ||\vec{x}|| ~ ||\vec{y}|| ~ cos(\theta)$$

Si on utilise les vecteurs unitaires correspondants à `x` et `y`, alors le module est à 1. On peut alors utiliser :

## 4 - Théorie: Interactivité avec `iwidget`

Il existe une fonction chouette nous permet de transformer n'importe quelle fonction python, en ***widget* interactif**, qu'on peut manipuler avec la souris.

On importe la fonction `interact`, puis on lui donne en paramètre le *nom* de la fonction avec laquelle on veut interagir, __sans parenthèses__, puis chaque paramètre est défini avec :

- Un intervalle de nombres avec des parenthèses `(minimum, maximum)`
- Un tableau de choix `["abc", "def"]`
- Une string, pour entrer du texte arbitraire


Il y a beaucoup d'autres options si vous fouillez dans la documentation, mais ça devrait déjà couvrir beaucoup de cas

```python
# IMPORTANT: Importer pour que la fonction interact() sois disponible
from ipywidgets import interact
```

```python
def presenter_personne(nom, apprecie, nb_fois):
  """Présente une personne un certain nombre de fois"""
  phrase = "Bonjour! Je m'appelle " + nom + " et j'apprécie " + apprecie

  texte = ""

  for i in range(nb_fois):
    texte += phrase + '\n'

  print(texte)

interact(presenter_personne,
    nom=["Alice", "Bob"],
    nb_fois=(1, 10),
    apprecie="les biscuits",
)
```

On peut se servir de `interact()` pour afficher de façon interactive des graphiques

```python
# Voici une fonction pour créer un petit graphique qui visualise deux vecteurs 2D :
import matplotlib.pyplot as plt

def afficher_vecteurs2D(x1, y1, x2, y2, couleur):
  """Prend deux vecteurs (x1, y1) et (x2, y2) et les affiche d'une certaine couleur.
  Exemples de couleurs: "r" pour rouge, "g" pour vert, "b" pour bleu

  Source - https://stackoverflow.com/a
  Posted by fuglede
  Retrieved 2025-11-14, License - CC BY-SA 3.0
  """

  plt.quiver([0, 0], [0,0], [x1, x2], [y1, y2], angles='xy', scale_units='xy', color=[couleur, couleur], scale=1)
  plt.xlim(-10, 10)
  plt.ylim(-10, 10)
  plt.show()

# Exemple: afficher_vecteurs2D(0, 5, 2, 4, 'r')
```

```python
interact(afficher_vecteurs2D,
    x1=(-10, 10),
    y1=(-10, 10),
    x2=(-10, 10),
    y2=(-10, 10),
    couleur=['r', 'g', 'b'],
)
```
