### 1.1 Une fonction qui prend en paramètre un texte, qui utilise `texte.split()` pour découper ce texte en mots et qui retourne un tableau avec seulement les mots de plus de 3 lettres ou plus

````python
#@title Exercice 1.1

def filtrer_long_mots(text):
  ...

###########

# Solution:
def filtrer_long_mots(text):
    mots_long = []
    mots = text.split()  # Découpe le texte en mots
    for mot in mots:
        if len(mot) > 3:  # Vérifie si le mot a plus de 3 lettres
            mots_long.append(mot)  # Ajoute le mot à la liste
    return mots_long

##### tests

texte = """L'apprentissage automatique comporte généralement deux phases. La première consiste à estimer un modèle à partir de données, appelées observations, qui sont disponibles et en nombre fini, lors de la phase de conception du système. L'estimation du modèle consiste à résoudre une tâche pratique, telle que traduire un discours, estimer une densité de probabilité, reconnaître la présence d'un chat dans une photographie ou participer à la conduite d'un véhicule autonome. Cette phase dite « d'apprentissage » ou « d'entraînement » est généralement préalable à l'utilisation pratique du modèle. La seconde phase est la mise en production : le modèle étant déterminé, de nouvelles données peuvent alors être soumises afin d'obtenir le résultat correspondant à la tâche souhaitée."""
print(filtrer_long_mots(texte))
````

### 1.2 Une fonction qui prend en paramètre un tableau et qui renvoie `True` si le tableau est symétrique, `False` sinon

Par exemple, `["a", "b", "b", "a"]` est symétrique car ce tableau serait identique si on le regardait à l'envers

````python
#@title Exercice 1.2
def est_symetrique(tableau):
  ...

######

def est_symetrique(tableau):
    longueur = len(tableau)
    for i in range(longueur // 2):
        if tableau[i] != tableau[longueur - 1 - i]:
            return False
    return True

#### tests
print(est_symetrique(["a", "b", "b", "a"]))
print(est_symetrique(["a", "b", "a"]))
print(est_symetrique(["a", "b", "a", "b"]))
print(est_symetrique(["a", "b", "b"]))
print(est_symetrique(["a"]))
print(est_symetrique([]))
````


### 1.3 La même fonction qu'au numéro précédent, mais qui ignore les espaces en début/fin et qui ignore la casse

Contrairement au numéro précédent, `["a", "b", "  b   ", "A"]` serait considéré comme symétrique ici

````python
#@title Exercice 1.3

def est_symetrique_a_peu_pres(tableau):
  ...

########

def est_symetrique_a_peu_pres(tableau):
    # On crée un second tableau, qui contient tous les mots "nettoyés"
    nouveau_tableau = []
    for mot in tableau:
        # .strip() enlève les espaces en début/fin
        # .lower() met en minuscules
        # Notez le fait qu'on appelle les deux une à la suite de l'autre ici
        # C'est équivalent à faire :
        #     mot = mot.strip()
        #     mot = mot.lower()
        # Mais d'un seul coup
        nouveau_tableau.append(mot.strip().lower())
    # On réutilise l'autre fonction telle quelle avec le nouveau tableau
    return est_symetrique(nouveau_tableau)

#### tests
print(est_symetrique_a_peu_pres(["a", "B", "b", "   a"]))
print(est_symetrique_a_peu_pres(["a", "b", "A"]))
print(est_symetrique_a_peu_pres(["a", "b   ", "a", "b"]))
print(est_symetrique_a_peu_pres(["A", " b ", "b"]))
print(est_symetrique_a_peu_pres([" a "]))
print(est_symetrique_a_peu_pres([]))
````


### 1.5 Une fonction qui calcule la distance entre deux points 2D $\vec{p_1} = (x_1, y_1)$ et $\vec{p_2} = (x_2, y_2)$, avez la formule

$$dist(\vec{p_1}, \vec{p_2}) = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}$$

````python
#@title Exercice 1.5
def distance(p1, p2):
  ...

########

def distance(p1, p2):
    x1, y1 = p1[0], p1[1]
    x2, y2 = p2[0], p2[1]
    return math.sqrt((x1 - x2)**2 + (y1 - y2)**2)

##### tests
p1 = [10, 15]
p2 = [5, 15]
p3 = [10, 25]
print(distance(p1, p2))
print(distance(p1, p3))
print(distance(p2, p3))
print(distance(p1, p1))
````


### 1.6 Une fonction qui prend en paramètre un seul tableau de points 2D et qui retourne un nouveau tableau de 2 éléments contenant la paire de points qui sont les plus proches l'un de l'autre

Par exemple, avec les 4 points `[[1, 2], [10, 20], [90, 100], [1, 3]]`

On devrait retourner un tableau qui contient seulement les 2: `[[1, 2], [1, 3]]`

````python
#@title Exercice 1.6
def points_les_plus_proches(points):
  ...

#########

def points_les_plus_proches(points):
  distance_min = 99999999999 # Nombre arbitrairement grand
  points_plus_proches = [points[0], points[1]]
  for i in range(len(points)):
    p1 = points[i]

    for j in range(i + 1, len(points)):

      if i == j:
        continue # On ne doit pas considérer un point avec lui-même

      p2 = points[j]

      dist = distance(p1, p2)

      if dist < distance_min:
        distance_min = dist
        points_plus_proches = [p1, p2]

  return points_plus_proches

####### tests
points = [
    [ 6, 45],
    [64, 95],
    [ 5, 52],
    [69, 72],
    [26, 30],
    [86, 89],
    [70, 63],
]
print(points_les_plus_proches(points))
````

En quoi est-ce que volontairement déclencher des erreurs peut-il nous être utile?

Parfois, mieux vaut que le code plante *tôt* quand la raison du plantage est claire, plutôt que *tard* quand on a déjà calculé un paquet de choses avec des variables qui ne faisaient pas de sens.

Supposons qu'on a la séquence de fonctions suivante :

````python
def addition_tableaux(u, v):
  """Additionne deux tableaux de dimensions quelconques"""

  somme = []

  for i in range(len(u)):
    somme.append(u[i] + v[i])

  return somme


def total(v):
  resultat = 0
  for x in v:
    resultat += x
  return resultat

def carre(x):
  return x*x


def faire_le_calcul():
  # Ça devrait faire [1+3, 2+4, 0+5] = [4, 6, 5]
  somme_vecteurs = addition_tableaux([1,2], [3,4,5])

  total_carre = carre(total(somme_vecteurs))

  # Ça devrait faire : 15² = 225
  return total_carre

# ... Pourquoi ça donne la mauvaise réponse???
print(faire_le_calcul())
````

Où est l'erreur?

Dans `addition_tableaux`, dans `total` ou dans `carre`?

Une des fonction faisait une hypothèse incorrecte avec ses données : `addition_vec` présume que les deux vecteurs auront la même taille, ce qui n'est pas le cas dans l'exemple d'utilisation de la méthode.

**`assert` permet de rendre explicite nos hypothèses de départ, en s'assurant que le code ne continue pas avec des données invalides si on se trompe.**

Même chose, avec un petit assert en début de fonction :

````python
def addition_tableaux(u, v):
  """Additionne deux tableaux de dimensions quelconques"""

  assert len(u) == len(v), "Les tableaux doivent avoir la même taille"

  somme = []

  for i in range(len(u)):
    somme.append(u[i] + v[i])

  return somme


def total(v):
  resultat = 0
  for x in v:
    resultat += x
  return resultat

def carre(x):
  return x*x


def faire_le_calcul():
  # Ça devrait faire [1+3, 2+4, 0+5] = [4, 6, 5]
  somme_vecteurs = addition_tableaux([1,2], [3,4,5])

  total_carre = carre(total(somme_vecteurs))

  # Ça devrait faire : 15² = 225
  return total_carre

### ******* DÉCOMMENTER POUR TESTER ******* ###
# print(faire_le_calcul())
````

### 2.1 le `produit_scalaire(u, v)`, qui retourne le produit scalaire (*dot product*) entre deux vecteurs à N dimensions

En 2D, le produit scalaire est :

$$(x, y) \cdot (a, b) = xa + yb$$

En 3D, ça serait plutôt :

$$||(x, y, z)|| = (x, y, z) \cdot (a, b, c) = xa + yb + zc$$

Généralisez la logique à n'importe quel nombre de dimensions.

````python
#@title Exercice 2.1
def produit_scalaire(u, v):
  assert len(u) == len(v), "Les vecteurs doivent être de même taille"
  resultat = 0
  for i in range(len(u)):
    resultat += u[i] * v[i]
  return resultat

assert produit_scalaire([1, 2, 3], [4, 5, 6]) == 32
assert produit_scalaire([5, 5], [1, 1]) == 10
````


### 2.2 le `module(vec)` pour un vecteur à N dimensions

Le module en 2D est :

$$||(x, y)|| = \sqrt{x^2 + y^2}$$

En 3D ça serait :

$$\sqrt{x^2 + y^2 + z^2}$$

En n'importe quel nombre de dimensions, ça serait :

$$\sqrt{x^2 + y^2 + ... + w^2}$$

````python
#@title Exercice 2.2
def module(vec):
  resultat = 0
  for x in vec:
    resultat += x*x
  return math.sqrt(resultat)

assert module([1, 0, 0]) == 1
assert module([10, 0, 0]) == 10
assert module([5, 5]) == distance([5, 5], [0, 0])
````


### 2.3 `vecteur_unitaire(v)`, qui retourne le vecteur où chaque élément est divisé par le module du vecteur

$$\vec{v}_\text{unitaire} = \frac{\vec{v}}{|| \vec{v} ||} = (\frac{v_0}{|| \vec{v} ||}, \frac{v_1}{|| \vec{v} ||}, ..., \frac{v_{N-1}}{|| \vec{v} ||}, )$$

````python
#@title Exercice 2.3
def unitaire(vec):
  resultat = []
  mod = module(vec)

  assert mod != 0, "On ne peut pas faire un vecteur unitaire de module zéro"

  for x in vec:
    resultat.append(x / mod)

  # Vérification de la réponse attendue
  assert round(module(resultat), 3) == 1, "Erreur dans le code: le module n'est pas à 1"

  return resultat

assert unitaire([1, 0, 0, 0]) == [1, 0, 0, 0]
assert unitaire([50, 0, 0, 0]) == [1, 0, 0, 0]
assert unitaire([4, 4, 4, 4]) == [1/2, 1/2, 1/2, 1/2]
````


### 2.4 `angle_entre_vecteurs(u, v)`

Le produit scalaire est défini avec :

$$\vec{x} \cdot \vec{y} = ||\vec{x}|| ~ ||\vec{y}|| ~ cos(\theta)$$

Où $\theta$ est l'angle entre les deux vecteurs et $||\vec{v}||$ est le module du vecteur.

Si on utilise les vecteurs unitaires correspondants à `x` et `y`, alors le module est à 1. On peut alors utiliser :

$$acos(\vec{x} \cdot \vec{y})$$

````python
#@title Exercice 2.4
def angle_entre_vecteurs_unitaires(u, v):
  # On réutilise les fonctions codées plus haut
  assert round(module(u), 3) == 1
  assert round(module(v), 3) == 1

  return math.degrees(math.acos(produit_scalaire(u, v)))

assert angle_entre_vecteurs_unitaires([1, 0], [0, 1]) == 90.0
assert angle_entre_vecteurs_unitaires([1, 0], [-1, 0]) == 180.0
assert angle_entre_vecteurs_unitaires([1, 0], [0, -1]) == 90.0
assert angle_entre_vecteurs_unitaires([1, 0], [1, 0]) == 0.0
assert round(angle_entre_vecteurs_unitaires([1, 0], unitaire([1, 1])), 3) == 45.0
assert round(angle_entre_vecteurs_unitaires(unitaire([1, -1]), unitaire([1, 1])), 3) == 90
````

## 4 - Théorie: Interactivité avec `iwidget`

Il existe une fonction chouette nous permet de transformer n'importe quelle fonction python, en ***widget* interactif**, qu'on peut manipuler avec la souris.

On importe la fonction `interact`, puis on lui donne en paramètre le *nom* de la fonction avec laquelle on veut interagir, __sans parenthèses__, puis chaque paramètre est défini avec :

- Un intervalle de nombres avec des parenthèses `(minimum, maximum)`
- Un tableau de choix `["abc", "def"]`
- Une string, pour entrer du texte arbitraire


Il y a beaucoup d'autres options si vous fouillez dans la documentation, mais ça devrait déjà couvrir beaucoup de cas

````python
#@title Importation
from ipywidgets import interact
````

````python
def presenter_personne(nom, apprecie, nb_fois):
  """Présente une personne un certain nombre de fois"""
  phrase = "Bonjour! Je m'appelle " + nom + " et j'apprécie " + apprecie

  texte = ""

  for i in range(nb_fois):
    texte += phrase + '\n'

  print(texte)

interact(presenter_personne,
    nom=["Alice", "Bob"],
    nb_fois=(1, 10),
    apprecie="les biscuits",
)
````

On peut se servir de `interact()` pour afficher de façon interactive des graphiques

````python
# Voici une fonction pour créer un petit graphique qui visualise deux vecteurs 2D :
import matplotlib.pyplot as plt

def afficher_vecteurs2D(x1, y1, x2, y2, couleur):
  """Prend deux vecteurs (x1, y1) et (x2, y2) et les affiche d'une certaine couleur.
  Exemples de couleurs: "r" pour rouge, "g" pour vert, "b" pour bleu

  Source - https://stackoverflow.com/a/42283226
  Posted by fuglede
  Retrieved 2025-11-14, License - CC BY-SA 3.0
  """

  plt.quiver([0, 0], [0,0], [x1, x2], [y1, y2], angles='xy', scale_units='xy', color=[couleur, couleur], scale=1)
  plt.xlim(-10, 10)
  plt.ylim(-10, 10)
  plt.show()

# Exemple non-interactif:
# afficher_vecteurs2D(0, 5, 2, 4, 'r')
````

````python
interact(afficher_vecteurs2D,
    x1=(-10, 10),
    y1=(-10, 10),
    x2=(-10, 10),
    y2=(-10, 10),
    couleur=['r', 'g', 'b'],
)
````
